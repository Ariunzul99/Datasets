# 1. Importe & Globale Konfiguration

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import StratifiedKFold, cross_validate, train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, roc_curve, confusion_matrix, ConfusionMatrixDisplay, RocCurveDisplay
)
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

RANDOM_STATE = 42
```

# 2. Datenvorverarbeitung, Cross-Validation & Metriken

```python
def evaluate_model_cv(X, y, model_pipeline, model_name, cv=None, scoring=None):
    if cv is None:
        cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
    if scoring is None:
        scoring = ['accuracy', 'precision', 'recall', 'f1', 'roc_auc']
    results = cross_validate(model_pipeline, X, y, cv=cv, scoring=scoring, return_train_score=False)
    summary = {}
    for metric in scoring:
        mean = np.mean(results[f'test_{metric}'])
        std = np.std(results[f'test_{metric}'])
        summary[metric] = f"{mean:.3f} ± {std:.3f}"
    print(f"{model_name}:")
    for k, v in summary.items():
        print(f"{k}: {v}")
    return results, summary
```

# 3. Evaluationsfunktion (ohne Data Leakage, mit CV + Holdout-Plots)

```python
def plot_holdout_results(model_pipeline, X, y, model_name):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=RANDOM_STATE)
    model_pipeline.fit(X_train, y_train)
    y_pred = model_pipeline.predict(X_test)
    y_prob = model_pipeline.predict_proba(X_test)[:, 1]

    print(f"\n{model_name} — Confusion Matrix (Holdout):")
    ConfusionMatrixDisplay(confusion_matrix(y_test, y_pred)).plot()
    plt.title(f"{model_name} Confusion Matrix (Holdout)")
    plt.show()

    RocCurveDisplay.from_predictions(y_test, y_prob)
    plt.title(f"{model_name} ROC Curve (Holdout)")
    plt.show()
```

# 4. Laden der German-Credit-Daten

```python
german_df = pd.read_csv("german_credit.csv")  # Pfad ggf. anpassen
for col in german_df.columns:
    if german_df[col].dtype == 'object':
        german_df[col] = german_df[col].astype('category').cat.codes
```

# 5. Zielvariable und Features definieren — German Credit

```python
y_german = (german_df["CreditRisk"] == 2).astype(int) # 1 = high risk, 0 = low risk
X_german = german_df.drop("CreditRisk", axis=1)
print("Zielvariable — Verteilung:", y_german.value_counts())
```

# 6. Pipelines aufbauen — German Credit (Logistische Regression & Random Forest)

```python
rf_g_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('rf', RandomForestClassifier(class_weight="balanced", random_state=RANDOM_STATE))
])

lr_g_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('lr', LogisticRegression(class_weight="balanced", max_iter=5000, random_state=RANDOM_STATE))
])
```

# 7. Modelle evaluieren — German Credit

```python
results_rf_g, summary_rf_g = evaluate_model_cv(X_german, y_german, rf_g_pipe, "Random Forest")
results_lr_g, summary_lr_g = evaluate_model_cv(X_german, y_german, lr_g_pipe, "Logistische Regression")

plot_holdout_results(rf_g_pipe, X_german, y_german, "Random Forest")
plot_holdout_results(lr_g_pipe, X_german, y_german, "Logistische Regression")
```

# 8. Koeffizienten der Logistischen Regression interpretieren — German Credit

```python
lr_g_pipe.fit(X_german, y_german)
feature_names = X_german.columns
lr_coef = lr_g_pipe.named_steps['lr'].coef_.ravel()
coef_df = pd.DataFrame({'Feature': feature_names, 'Koeffizient': lr_coef, 'Abs_Koeffizient': np.abs(lr_coef)})
coef_df_sorted = coef_df.sort_values('Abs_Koeffizient', ascending=False)
print(coef_df_sorted)
plt.figure(figsize=(8,4))
plt.barh(coef_df_sorted['Feature'], coef_df_sorted['Koeffizient'])
plt.xlabel("Koeffizient")
plt.title("LR-Koeffizienten — German Credit")
plt.show()
```

# 8a. Feature-Importances des Random Forest interpretieren — German Credit

```python
rf_g_pipe.fit(X_german, y_german)
rf_importances = rf_g_pipe.named_steps['rf'].feature_importances_
rf_df = pd.DataFrame({'Feature': X_german.columns, 'Importance': rf_importances})
rf_df_sorted = rf_df.sort_values('Importance', ascending=False)
print(rf_df_sorted)
plt.figure(figsize=(8,4))
plt.barh(rf_df_sorted['Feature'], rf_df_sorted['Importance'])
plt.xlabel("Importance")
plt.title("RF Feature Importances — German Credit")
plt.show()
```

# 9. Laden der Credit-Card-Default-Daten (Taiwan)

```python
cc_df = pd.read_csv("default-of-credit-card-clients-2.csv")  # Pfad ggf. anpassen
cc_df = cc_df.drop(index=0).copy()
if 'Unnamed: 0' in cc_df.columns: cc_df = cc_df.drop('Unnamed: 0', axis=1)
if 'X1' in cc_df.columns: cc_df = cc_df.drop('X1', axis=1)
if 'Y' in cc_df.columns: cc_df = cc_df.rename(columns={'Y': 'target'})
cc_df = cc_df.apply(pd.to_numeric, errors='coerce')
cc_df = cc_df.dropna()
```

# 10. Zielvariable und Features definieren — Taiwan Credit

```python
y_cc = cc_df['target'].astype(int)
X_cc = cc_df.drop(columns=['target'])
print("Zielvariable — Verteilung:", y_cc.value_counts())
```

# 11. Pipelines aufbauen — Taiwan Credit (Logistische Regression & Random Forest)

```python
rf_c_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('rf', RandomForestClassifier(class_weight="balanced", random_state=RANDOM_STATE))
])

lr_c_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('lr', LogisticRegression(class_weight="balanced", max_iter=5000, random_state=RANDOM_STATE))
])
```

# 12. Modelle evaluieren — Taiwan Credit

```python
results_rf_c, summary_rf_c = evaluate_model_cv(X_cc, y_cc, rf_c_pipe, "Random Forest")
results_lr_c, summary_lr_c = evaluate_model_cv(X_cc, y_cc, lr_c_pipe, "Logistische Regression")

plot_holdout_results(rf_c_pipe, X_cc, y_cc, "Random Forest")
plot_holdout_results(lr_c_pipe, X_cc, y_cc, "Logistische Regression")
```

# 13. Koeffizienten der Logistischen Regression interpretieren — Taiwan Credit

```python
lr_c_pipe.fit(X_cc, y_cc)
feature_names_cc = X_cc.columns
lr_coef_cc = lr_c_pipe.named_steps['lr'].coef_.ravel()
coef_df_cc = pd.DataFrame({'Feature': feature_names_cc, 'Koeffizient': lr_coef_cc, 'Abs_Koeffizient': np.abs(lr_coef_cc)})
coef_df_cc_sorted = coef_df_cc.sort_values('Abs_Koeffizient', ascending=False)
print(coef_df_cc_sorted)
plt.figure(figsize=(8,4))
plt.barh(coef_df_cc_sorted['Feature'], coef_df_cc_sorted['Koeffizient'])
plt.xlabel("Koeffizient")
plt.title("LR-Koeffizienten — Taiwan Credit")
plt.show()
```

# 13a. Feature-Importances des Random Forest interpretieren — Taiwan Credit

```python
rf_c_pipe.fit(X_cc, y_cc)
rf_importances_cc = rf_c_pipe.named_steps['rf'].feature_importances_
rf_df_cc = pd.DataFrame({'Feature': X_cc.columns, 'Importance': rf_importances_cc})
rf_df_cc_sorted = rf_df_cc.sort_values('Importance', ascending=False)
print(rf_df_cc_sorted)
plt.figure(figsize=(8,4))
plt.barh(rf_df_cc_sorted['Feature'], rf_df_cc_sorted['Importance'])
plt.xlabel("Importance")
plt.title("RF Feature Importances — Taiwan Credit")
plt.show()
```

# 14. Zusammenfassungstabelle — Cross-Validated Kennzahlen (Mittelwert ± Std) 

```python
summary_german = pd.DataFrame([summary_rf_g, summary_lr_g], index=['Random Forest (German)', 'Logistische Regression (German)'])
summary_taiwan = pd.DataFrame([summary_rf_c, summary_lr_c], index=['Random Forest (Taiwan)', 'Logistische Regression (Taiwan)'])
summary_all = pd.concat([summary_german, summary_taiwan])
print(summary_all)

plt.figure(figsize=(20, 8))  # Extra large figure for readability
plt.title("Cross-Validated Kennzahlen (Mittelwert ± Std)", fontsize=22)
plt.axis('off')
tbl = plt.table(cellText=summary_all.values,
                rowLabels=summary_all.index,
                colLabels=summary_all.columns,
                cellLoc='center',
                loc='center',
                bbox=[0, 0, 1, 1])  # Occupy full figure
tbl.auto_set_font_size(False)
tbl.set_fontsize(18)  # Larger font for visibility
tbl.scale(2, 2.2)    # Widen and heighten table
plt.show()
```
