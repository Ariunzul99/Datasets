```python
# Import essential libraries for ML pipeline
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
                             roc_auc_score, roc_curve, confusion_matrix, ConfusionMatrixDisplay)
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
# Function to evaluate model performance with comprehensive metrics
def evaluate_model(X, y, model, model_name, dataset_name):
    # Split data: 80% training, 20% testing (stratified to maintain class balance)
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, stratify=y, random_state=42
    )
    
    # Train model and make predictions
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)  # Binary predictions
    y_prob = model.predict_proba(X_test)[:, 1]  # Probability scores for ROC
    
    # Calculate key performance metrics
    acc = accuracy_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_prob)
    
    print(f"\n{dataset_name} - {model_name}")
    print(f"Accuracy: {acc:.4f}")
    print(f"ROC AUC: {roc_auc:.4f}")
    
    # Display confusion matrix (actual vs predicted)
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.title(f"Confusion Matrix: {model_name} ({dataset_name})")
    plt.show()
    
    # Cross-validation: Test model on 5 different train/test splits
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    cv_acc = cross_val_score(model, X, y, cv=skf, scoring='accuracy')
    cv_auc = cross_val_score(model, X, y, cv=skf, scoring='roc_auc')
    print(f"Cross-validated Accuracy (mean): {cv_acc.mean():.4f}")
    print(f"Cross-validated ROC AUC (mean): {cv_auc.mean():.4f}")
    
    # Show feature importance (Random Forest only)
    if hasattr(model, "feature_importances_"):
        # Handle both DataFrame and numpy array inputs
        if isinstance(X, np.ndarray):
            feature_names = [f"X{i}" for i in range(X.shape[1])]
        else:
            feature_names = X.columns
        
        # Plot top features that drive predictions
        importances = pd.Series(model.feature_importances_, index=feature_names)
        importances.sort_values(ascending=False).plot(kind='bar', figsize=(10,4))
        plt.title(f"Feature Importance: {model_name} ({dataset_name})")
        plt.tight_layout()
        plt.show()
    
    # Calculate ROC curve data for comparison plots
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    return acc, roc_auc, cv_acc.mean(), cv_auc.mean(), fpr, tpr
```

```python
# ----------- German Credit Data -----------
print("="*50)
print("GERMAN CREDIT DATA ANALYSIS")
print("="*50)

# Load German credit risk dataset
german_df = pd.read_csv("german_credit.csv")

# Encode categorical variables to numerical format (required for ML algorithms)
for col in german_df.columns:
    if german_df[col].dtype == 'object':
        german_df[col] = german_df[col].astype('category').cat.codes

# Prepare features and target variable
y_german = (german_df["CreditRisk"] == 2).astype(int)  # Convert to binary: 1=high risk, 0=low risk
X_german = german_df.drop("CreditRisk", axis=1)  # Remove target from feature set

print("Value counts for y_german:")
print(y_german.value_counts())  # Check class distribution
print(f"Dataset shape: {german_df.shape}")
print(f"Features: {X_german.shape[1]}")

# Standardize features: mean=0, std=1 (essential for Logistic Regression)
scaler_g = StandardScaler()
X_german_scaled = scaler_g.fit_transform(X_german)

# Initialize models with reproducible random seeds
rf_g = RandomForestClassifier(random_state=42)  # Ensemble of decision trees
lr_g = LogisticRegression(max_iter=5000, random_state=42)  # Linear probability model

# Enhanced evaluation function with detailed metrics
def evaluate_german_model(X, y, model, model_name):
    try:
        print(f"\n{'='*30}")
        print(f"Starting {model_name} Analysis...")
        print(f"{'='*30}")
        
        # Standard train-test split
        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=0.2, stratify=y, random_state=42
        )
        
        print(f"Fitting {model_name}...")
        model.fit(X_train, y_train)
        
        print(f"Making predictions...")
        y_pred = model.predict(X_test)
        y_prob = model.predict_proba(X_test)[:, 1]
        
        # Calculate comprehensive performance metrics
        acc = accuracy_score(y_test, y_pred)  # Overall correctness
        precision = precision_score(y_test, y_pred, zero_division=0)  # True positives / Predicted positives
        recall = recall_score(y_test, y_pred, zero_division=0)  # True positives / Actual positives
        f1 = f1_score(y_test, y_pred, zero_division=0)  # Harmonic mean of precision & recall
        roc_auc = roc_auc_score(y_test, y_prob)  # Discrimination ability (0.5=random, 1.0=perfect)
        
        print(f"\nGerman Credit - {model_name}")
        print(f"Accuracy: {acc:.4f}")
        print(f"Precision: {precision:.4f}")
        print(f"Recall: {recall:.4f}")
        print(f"F1-Score: {f1:.4f}")
        print(f"ROC AUC: {roc_auc:.4f}")
        
        # Visual confusion matrix (actual vs predicted)
        cm = confusion_matrix(y_test, y_pred)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm)
        disp.plot()
        plt.title(f"Confusion Matrix: {model_name} - German Credit")
        plt.show()
        
        # Cross-validation: Test on 5 different train/test splits for robust estimates
        print(f"Running cross-validation for {model_name}...")
        skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
        cv_acc = cross_val_score(model, X, y, cv=skf, scoring='accuracy')
        cv_auc = cross_val_score(model, X, y, cv=skf, scoring='roc_auc')
        print(f"Cross-validated Accuracy (mean): {cv_acc.mean():.4f}")
        print(f"Cross-validated ROC AUC (mean): {cv_auc.mean():.4f}")
        
        # Feature importance visualization (Random Forest only)
        if hasattr(model, "feature_importances_"):
            print(f"Plotting feature importances...")
            feature_names = X_german.columns if hasattr(X_german, 'columns') else [f"Feature_{i}" for i in range(X.shape[1])]
            importances = pd.Series(model.feature_importances_, index=feature_names)
            top_features = importances.sort_values(ascending=False).head(10)
            
            plt.figure(figsize=(10, 6))
            top_features.plot(kind='bar')
            plt.title(f"Top 10 Feature Importances: {model_name} - German Credit")
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.show()
        
        # Store ROC curve data for comparison plots
        fpr, tpr, _ = roc_curve(y_test, y_prob)
        
        print(f"{model_name} analysis completed successfully!")
        
        return {
            'accuracy': acc, 'precision': precision, 'recall': recall, 'f1': f1, 'roc_auc': roc_auc,
            'cv_acc': cv_acc.mean(), 'cv_auc': cv_auc.mean(), 'fpr': fpr, 'tpr': tpr
        }
        
    except Exception as e:
        print(f"Error in {model_name}: {str(e)}")
        return None

# Execute analysis for both models
print("Starting Random Forest Analysis...")
results_german_rf = evaluate_german_model(X_german_scaled, y_german, rf_g, "Random Forest")

print("\nStarting Logistic Regression Analysis...")
results_german_lr = evaluate_german_model(X_german_scaled, y_german, lr_g, "Logistic Regression")

# Model comparison and visualization
if results_german_rf is not None and results_german_lr is not None:
    # Create side-by-side comparison table
    german_results_df = pd.DataFrame([
        {'Model': 'Random Forest', **{k: v for k, v in results_german_rf.items() if k not in ['fpr', 'tpr']}},
        {'Model': 'Logistic Regression', **{k: v for k, v in results_german_lr.items() if k not in ['fpr', 'tpr']}}
    ])

    print("\n" + "="*50)
    print("GERMAN CREDIT - MODEL COMPARISON TABLE")
    print("="*50)
    print(german_results_df.round(4))

    # ROC Curves: Visual comparison of model discrimination ability
    plt.figure(figsize=(10, 6))
    plt.plot(results_german_rf['fpr'], results_german_rf['tpr'], 
             label=f'Random Forest (AUC={results_german_rf["roc_auc"]:.3f})', linewidth=2)
    plt.plot(results_german_lr['fpr'], results_german_lr['tpr'], 
             label=f'Logistic Regression (AUC={results_german_lr["roc_auc"]:.3f})', linewidth=2)
    plt.plot([0,1],[0,1],'k--', alpha=0.5, label='Random Classifier')  # Baseline (no skill)
    plt.title("ROC Curves - German Credit Risk Prediction")
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(loc='lower right')
    plt.grid(True, alpha=0.3)
    plt.show()

    print("German Credit analysis completed for both models!")
else:
    print("One or both models failed to complete. Check the error messages above.")

# INTERPRETABILITY ANALYSIS: Logistic Regression Coefficients
if results_german_lr is not None:
    print("\n" + "="*50)
    print("GERMAN CREDIT - LOGISTIC REGRESSION COEFFICIENTS (Standardized Scale)")
    print("="*50)
    
    # Refit model to extract coefficients
    X_train, X_test, y_train, y_test = train_test_split(
        X_german_scaled, y_german, test_size=0.2, stratify=y_german, random_state=42
    )
    lr_fitted = LogisticRegression(max_iter=5000, random_state=42)
    lr_fitted.fit(X_train, y_train)
    
    # Extract and analyze feature coefficients
    feature_names = X_german.columns if hasattr(X_german, 'columns') else [f"Feature_{i}" for i in range(X_german_scaled.shape[1])]
    coefficients = pd.Series(lr_fitted.coef_.ravel(), index=feature_names)
    
    # Sort by absolute importance (magnitude of impact)
    coefficients_sorted = coefficients.sort_values(key=abs, ascending=False)
    
    print("German Credit - Logistic Regression Coefficients (sorted by absolute importance):")
    print(coefficients_sorted.round(6))
    
    # Visualize coefficients: positive=increases risk, negative=decreases risk
    plt.figure(figsize=(12, 8))
    coefficients_sorted.plot(kind='barh')
    plt.title('Logistic Regression Coefficients - German Credit Risk')
    plt.xlabel('Coefficient Value (Standardized Scale)')
    plt.ylabel('Features')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    # Business interpretation: Top risk factors
    print(f"\nTop 5 Positive Coefficients (increase credit risk probability):")
    positive_coefs = coefficients_sorted[coefficients_sorted > 0].head(5)
    for feature, coef in positive_coefs.items():
        print(f"  {feature}: {coef:.6f}")
    
    print(f"\nTop 5 Negative Coefficients (decrease credit risk probability):")
    negative_coefs = coefficients_sorted[coefficients_sorted < 0].head(5)
    for feature, coef in negative_coefs.items():
        print(f"  {feature}: {coef:.6f}")
    
    # Feature importance comparison: Random Forest vs Logistic Regression
    if results_german_rf is not None:
        print("\n" + "="*40)
        print("GERMAN CREDIT - FEATURE IMPORTANCE COMPARISON")
        print("="*40)
        
        # Refit Random Forest to get feature importances
        rf_fitted = RandomForestClassifier(random_state=42)
        rf_fitted.fit(X_train, y_train)
        
        rf_importance = pd.Series(rf_fitted.feature_importances_, index=feature_names)
        lr_importance = coefficients.abs()  # Use absolute values for comparison
        
        # Side-by-side comparison of feature rankings
        comparison_df = pd.DataFrame({
            'Random_Forest_Importance': rf_importance,
            'Logistic_Regression_AbsCoeff': lr_importance,
            'Logistic_Regression_Coeff': coefficients  # Original coefficients with sign
        })
        
        # Sort by Random Forest importance
        comparison_df = comparison_df.sort_values('Random_Forest_Importance', ascending=False)
        
        print("Top 10 Features Comparison:")
        print(comparison_df.head(10).round(6))
        
        # Visual comparison of top 10 features
        top_10_features = comparison_df.head(10)
        
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
        
        # Random Forest: Impurity-based importance
        top_10_features['Random_Forest_Importance'].plot(kind='barh', ax=ax1)
        ax1.set_title('German Credit - Random Forest Feature Importance')
        ax1.set_xlabel('Importance')
        
        # Logistic Regression: Coefficient magnitude and direction
        top_10_features['Logistic_Regression_Coeff'].plot(kind='barh', ax=ax2)
        ax2.set_title('German Credit - Logistic Regression Coefficients')
        ax2.set_xlabel('Coefficient Value')
        ax2.axvline(x=0, color='black', linestyle='--', alpha=0.5)  # Zero reference line
        
        plt.tight_layout()
        plt.show()
        
        # Export results for further analysis
        coefficients_df = pd.DataFrame({
            'Feature': coefficients_sorted.index,
            'Coefficient': coefficients_sorted.values,
            'Abs_Coefficient': coefficients_sorted.abs().values
        })
        coefficients_df.to_csv("german_credit_logistic_coefficients.csv", index=False)
        print(f"\nGerman Credit logistic regression coefficients saved to german_credit_logistic_coefficients.csv")

else:
    print("Logistic Regression failed, cannot show coefficients")
```

```python
# ----------- Credit Card Default Data -----------
cc_df = pd.read_csv("default-of-credit-card-clients-2.csv")

# Data preprocessing: Remove header row and ID columns
cc_df = cc_df.drop(index=0).copy()  # First row contains duplicate headers

# Clean dataset by removing non-predictive columns
cols_to_drop = []
for col in ['Unnamed: 0', 'X1']:  # ID-like columns
    if col in cc_df.columns:
        cols_to_drop.append(col)
if cols_to_drop:
    cc_df = cc_df.drop(columns=cols_to_drop)

# Standardize target variable name
if 'Y' in cc_df.columns:
    cc_df = cc_df.rename(columns={'Y': 'target'})

# Convert all columns to numeric format
cc_df = cc_df.apply(pd.to_numeric, errors='coerce')
cc_df = cc_df.dropna()  # Remove any rows with missing values

# Prepare features and target
target_col_cc = 'target'
y_cc = cc_df[target_col_cc].astype(int)  # 1=default, 0=no default
X_cc = cc_df.drop(columns=[target_col_cc])

print("Value counts for y_cc:")
print(y_cc.value_counts())  # Check class distribution (imbalanced dataset)

# Feature scaling for optimal model performance
scaler_c = StandardScaler()
X_cc_scaled = scaler_c.fit_transform(X_cc)

# Initialize models
rf_c = RandomForestClassifier(random_state=42)
lr_c = LogisticRegression(max_iter=5000, random_state=42)

# Credit card specific evaluation function
def evaluate_credit_model(X, y, model, model_name):
    """Evaluate credit card default prediction models"""
    try:
        print(f"\n{'='*30}")
        print(f"Starting {model_name} Analysis...")
        print(f"{'='*30}")
        
        # Stratified split maintains class proportion in train/test sets
        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=0.2, stratify=y, random_state=42
        )
        
        print(f"Fitting {model_name}...")
        model.fit(X_train, y_train)
        
        print(f"Making predictions...")
        y_pred = model.predict(X_test)
        y_prob = model.predict_proba(X_test)[:, 1]
        
        # Comprehensive metric calculation
        acc = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, zero_division=0)
        recall = recall_score(y_test, y_pred, zero_division=0)
        f1 = f1_score(y_test, y_pred, zero_division=0)
        roc_auc = roc_auc_score(y_test, y_prob)
        
        print(f"\nCredit Card Default - {model_name}")
        print(f"Accuracy: {acc:.4f}")
        print(f"Precision: {precision:.4f}")
        print(f"Recall: {recall:.4f}")
        print(f"F1-Score: {f1:.4f}")
        print(f"ROC AUC: {roc_auc:.4f}")
        
        # Confusion matrix visualization
        cm = confusion_matrix(y_test, y_pred)
        disp = ConfusionMatrixDisplay(confusion_matrix=cm)
        disp.plot()
        plt.title(f"Confusion Matrix: {model_name} - Credit Card Default")
        plt.show()
        
        # Cross-validation for robust performance estimation
        print(f"Running cross-validation for {model_name}...")
        skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
        cv_acc = cross_val_score(model, X, y, cv=skf, scoring='accuracy')
        cv_auc = cross_val_score(model, X, y, cv=skf, scoring='roc_auc')
        print(f"Cross-validated Accuracy (mean): {cv_acc.mean():.4f}")
        print(f"Cross-validated ROC AUC (mean): {cv_auc.mean():.4f}")
        
        # Feature importance analysis (Random Forest only)
        if hasattr(model, "feature_importances_"):
            print(f"Plotting feature importances...")
            feature_names = [f"X{i}" for i in range(X.shape[1])] if isinstance(X, np.ndarray) else X.columns
            importances = pd.Series(model.feature_importances_, index=feature_names)
            top_features = importances.sort_values(ascending=False).head(10)
            
            plt.figure(figsize=(10, 6))
            top_features.plot(kind='bar')
            plt.title(f"Top 10 Feature Importances: {model_name} - Credit Card Default")
            plt.xticks(rotation=45)
            plt.tight_layout()
            plt.show()
        
        # ROC curve data for comparison
        fpr, tpr, _ = roc_curve(y_test, y_prob)
        
        print(f"{model_name} analysis completed successfully!")
        
        return {
            'accuracy': acc, 'precision': precision, 'recall': recall, 'f1': f1, 'roc_auc': roc_auc,
            'cv_acc': cv_acc.mean(), 'cv_auc': cv_auc.mean(), 'fpr': fpr, 'tpr': tpr
        }
        
    except Exception as e:
        print(f"Error in {model_name}: {str(e)}")
        return None

# Execute analysis for both models
print("Starting Random Forest Analysis...")
results_cc_rf = evaluate_credit_model(X_cc_scaled, y_cc, rf_c, "Random Forest")

print("\nStarting Logistic Regression Analysis...")
results_cc_lr = evaluate_credit_model(X_cc_scaled, y_cc, lr_c, "Logistic Regression")

# Model comparison and results summary
if results_cc_rf is not None and results_cc_lr is not None:
    # Performance comparison table
    cc_results_df = pd.DataFrame([
        {'Model': 'Random Forest', **{k: v for k, v in results_cc_rf.items() if k not in ['fpr', 'tpr']}},
        {'Model': 'Logistic Regression', **{k: v for k, v in results_cc_lr.items() if k not in ['fpr', 'tpr']}}
    ])

    print("\n" + "="*50)
    print("CREDIT CARD DEFAULT - MODEL COMPARISON TABLE")
    print("="*50)
    print(cc_results_df.round(4))

    # ROC curve comparison: Visual assessment of model discrimination
    plt.figure(figsize=(10, 6))
    plt.plot(results_cc_rf['fpr'], results_cc_rf['tpr'], 
             label=f'Random Forest (AUC={results_cc_rf["roc_auc"]:.3f})', linewidth=2)
    plt.plot(results_cc_lr['fpr'], results_cc_lr['tpr'], 
             label=f'Logistic Regression (AUC={results_cc_lr["roc_auc"]:.3f})', linewidth=2)
    plt.plot([0,1],[0,1],'k--', alpha=0.5, label='Random Classifier')
    plt.title("ROC Curves - Credit Card Default Prediction")
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend(loc='lower right')
    plt.grid(True, alpha=0.3)
    plt.show()

    print("Credit Card Default analysis completed for both models!")
else:
    print("One or both models failed to complete. Check the error messages above.")
    
# COEFFICIENT ANALYSIS: Understanding feature impact on default probability
if results_cc_lr is not None:
    print("\n" + "="*50)
    print("LOGISTIC REGRESSION COEFFICIENTS (Standardized Scale)")
    print("="*50)
    
    # Refit model to extract coefficients
    X_train, X_test, y_train, y_test = train_test_split(
        X_cc_scaled, y_cc, test_size=0.2, stratify=y_cc, random_state=42
    )
    lr_fitted = LogisticRegression(max_iter=5000, random_state=42)
    lr_fitted.fit(X_train, y_train)
    
    # Extract feature coefficients
    feature_names = [col for col in cc_df.columns if col != 'target']
    coefficients = pd.Series(lr_fitted.coef_.ravel(), index=feature_names)
    
    # Sort by absolute magnitude of impact
    coefficients_sorted = coefficients.sort_values(key=abs, ascending=False)
    
    print("Logistic Regression Coefficients (sorted by absolute importance):")
    print(coefficients_sorted.round(6))
    
    # Coefficient visualization: Positive increases default risk, negative decreases it
    plt.figure(figsize=(12, 8))
    coefficients_sorted.plot(kind='barh')
    plt.title('Logistic Regression Coefficients - Credit Card Default')
    plt.xlabel('Coefficient Value (Standardized Scale)')
    plt.ylabel('Features')
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    # Business insights: Key risk factors
    print(f"\nTop 5 Positive Coefficients (increase default probability):")
    positive_coefs = coefficients_sorted[coefficients_sorted > 0].head(5)
    for feature, coef in positive_coefs.items():
        print(f"  {feature}: {coef:.6f}")
    
    print(f"\nTop 5 Negative Coefficients (decrease default probability):")
    negative_coefs = coefficients_sorted[coefficients_sorted < 0].head(5)
    for feature, coef in negative_coefs.items():
        print(f"  {feature}: {coef:.6f}")

else:
    print("Logistic Regression failed, cannot show coefficients")
```
